```python
# Function to convert years and months to years
def convert_to_years(lease):
    # Use regular expressions to extract years and months
    years_match = re.search(r'(\d+) years?', lease)
    months_match = re.search(r'(\d+) months?', lease)
    
    years = int(years_match.group(1)) if years_match else 0
    months = int(months_match.group(1)) if months_match else 0
    
    # Convert months to a fraction of years
    months_in_years = months / 12
    
    return years + months_in_years

# Apply the function to the DataFrame
df['remaining_lease_years'] = df['remaining_lease'].apply(convert_to_years)

# Drop remaining_lease
df = df.drop('remaining_lease', axis =1)

# Display the DataFrame
print(df)
```

```python
# check skewness 
selected_var = ['floor_area_sqm','lease_commence_date','resale_price','remaining_lease_years']
skew = df[selected_var].skew()

print("Skewness:")
print(skew)

print("==========")

# check kurtosis 
kurt = scipy.stats.kurtosis(df[selected_var])

print("Kurtosis:")
print(selected_var[0],":\t", kurt[0])
print(selected_var[1],":\t", kurt[1])
print(selected_var[2],":\t\t", kurt[2])
print(selected_var[3],":\t", kurt[3])
```

Base on the data above, 
Skewness for resale price is 1.00 which is strongly positive skewed. This indicates that the 'resale_price' mean is much greater than the median and most of the observations are concentrated on the right tail.

'resale_price' of 1.06 has a light tail which indicates few extreme values/outliers. Therefore, the distribution is relatively close to a normal distribution.

'floor_area_sqm' kurtosis of -0.10 indicates shorter and thinner tails. The peak is lower and broader which fewer extreme values and a flatter graph

On the contary, 'lease_commence_date' and 'remaining_lease_years' 
 indicates a lower peak with lighter tails
